```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("iris.csv")
```

```python
df.fillna(method="bfill",inplace=True)
```

```python
df.isna().sum()
```

```python
from sklearn.feature_selection import SelectKBest,chi2,f_regression
```

```python
df.head()
```

```python
x=df.iloc[:,0:4]
y=df.Species #y=df.iloc[:,4] 
```

```python
x.head()
```

```python
y.head()
```

```python
selectk=SelectKBest(chi2,k=2)          #creating objects using parameterized constructor
#chi2 since y is classification problem
selectk.fit_transform(x,y)
```

```python
cols=selectk.get_support(indices=True) #taking index values of select k.
cols
```

```python
new_df=df.iloc[:,cols]
```

```python
new_df.head()
```

```python
selectk=SelectKBest(chi2,k=1)          #creating objects using parameterized constructor
#chi2 since y is classification problem
selectk.fit_transform(x,y)
```

```python
cols=selectk.get_support(indices=True) #taking index values of select k.
cols
```

```python
new_df=df.iloc[:,cols]
new_df.head()
```

```python
selectk=SelectKBest(chi2,k=3)          #creating objects using parameterized constructor
#chi2 since y is classification problem
selectk.fit_transform(x,y)
```

```python
cols=selectk.get_support(indices=True) #taking index values of select k.
cols
```

```python
new_df=df.iloc[:,cols]
new_df.head()
```
