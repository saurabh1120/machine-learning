##### MinMax scaling - uses normalization
##### standard scaling - user standardization
##### Binarizer - either 0 or 1 based on threshold 



```python
from sklearn.preprocessing import MinMaxScaler

mm=MinMaxScaler(feature_range=(1,4))

x=df4.iloc[:,0:4]

mm.fit_transform(x)
```

```python
from sklearn.preprocessing import StandardScaler

ss=StandardScaler()

x=df4.iloc[:,0:4]

ss.fit_transform(x)
```

```python
from sklearn.preprocessing import Binarizer

Br=Binarizer(threshold=2)

x=df4.iloc[:,0:4]

x=x.dropna()

Br.fit_transform(x)
```
