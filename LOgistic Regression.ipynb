### Importing Libaries and Read the Dataset

```python
import pandas as pd
import numpy as np
df=pd.read_csv("E:\\ML\\Housing.csv")
df.head()
```

# converting of price as a categorical Variable

```python
from sklearn.preprocessing import Binarizer
b=Binarizer(threshold=df['price'].mean())
y=b.fit_transform(df[['price']])
```

```python
x=df[['lotsize']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1,shuffle=True)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
ypred=lr.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,ypred)
```
